```python
# Import libraries
import os

import pandas as pd
from pandas.api.types import is_string_dtype, is_numeric_dtype
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
# Import earthquakes data
URL = 'https://raw.githubusercontent.com/carmengg/eds220-hwk-1/main/data/earthquake_data.csv'
earthquakes = pd.read_csv(URL)

# Import country border geodata
countries = gpd.read_file('data/ne_110m_admin_0_countries/ne_110m_admin_0_countries.shp')
```

```python
# Create points from latitude and longitude
points = gpd.points_from_xy(earthquakes.longitude, 
                           earthquakes.latitude)

# Create geodataframe
earthquake_points = gpd.GeoDataFrame(earthquakes,       # Data
                                     geometry=points,   # Specify geometry column
                                     crs=countries.crs  # Specify CRS
                                    )
```

```python
# Check to see if geometry column was added
earthquake_points.head()
```

```python
# Select all earthquake events in the `type` column
earthquake_points_subset = earthquake_points[earthquake_points['type'] == 'earthquake']
```

```python
fig, ax = plt.subplots(figsize=(8, 4))

# remove the axis for a cleaner map
ax.axis('off')

# Title for the plot
ax.set_title('Earthquakes of magnitude at least 6.5 in 2014',
             fontsize = 15)

# Add countries
countries.plot(ax=ax,
                color='lightgrey',
                edgecolor='none')

# Add earthquakes
earthquake_points_subset[earthquake_points_subset['mag'] >= 6.5 ].plot(ax=ax,
                column='mag',
                legend=True,
                markersize=28,
                cmap='Oranges',
                edgecolor='darkred',                                                   
                legend_kwds={
                     'label': 'moment magnitude'
                 })

# Add credits
plt.figtext(0.1, 0.1, "Data: USGS Earthquakes Database", fontsize=8, color="grey")
```
